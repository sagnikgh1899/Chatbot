```python
import numpy as np
import random
from scipy.spatial import distance
import math
import time

rows,columns = input("\nEnter the data dimension(for plotting purpose) in the form of rows (space) columns\n").split()
lof_value_by_me = 1.1

name_of_file = 'datafile.csv'
data = np.delete((np.genfromtxt(name_of_file,delimiter=',')),obj=0,axis=0)
data=data/np.linalg.norm(data)

import time
tim = time.time()

print(data,"\n\n")

lrd_of_o_for_LOF = [] #Very Important
outlier_list = []

##### Coding of the k distance; Status- Achieved;

tem1 = len(data)
print("Length =",tem1,"\n")
    
print("Enter the value of k between 1 and",(tem1-1),"(k stands for the 'kth' nearest neighbour value)")
k = int(input())

        
##### OBJECTIVE- TO FIND THE SET LAMBDA; STATUS- ACHIEVED;
        
a = 0
knn_list = []
dist_list = []
        
for j2 in range(0,tem1): 
    for i2 in range(0,tem1):
        dist = distance.euclidean(data[j2],data[i2])
        dist_list.append(dist)
    dist_list.sort(reverse=False)
    knn_list.append(dist_list[k])
    dist_list = []
    a = a + 1
            
#print("\nSet Lambda:- ",knn_list,"\n")  # We have obtained the set lambda  ##Turned off for convenience


#####  To calculate the set of N neighbours of p; Status- Achieved;
print("\n*********************************************************************\n")
#print("Enter the value of k between 1 and",(tem1-1))
#k = int(input())


whole = []
n_neigh_list = []
for j2 in range(0,tem1):
    for i2 in range(0,tem1):
        dist = distance.euclidean(data[j2],data[i2])
        whole.append([dist,data[j2],data[i2]])
    whole.sort(key=lambda x: x[0])
    #print(whole,"\n")
    for i3 in range(1,(k+1)):
        n_neigh_list.append(whole[i3][2])
    print("The",k,"nearest neighbours of",data[j2],"are:\n",n_neigh_list,"\n") 
    
    
    ##### To get the values of the k-distance of 'o'; Status- Achieved;
    len_n_neigh_list = len(n_neigh_list)
    
    whole1 = []
    n_neigh_of_o_list = []
    for i4 in range(0,len_n_neigh_list):
        for j4 in range(0,tem1):
            dist1 = distance.euclidean(n_neigh_list[i4],data[j4])
            whole1.append([dist1,n_neigh_list[i4],data[j4]])
        whole1.sort(key=lambda x: x[0])
        #print(whole1,"\n")   ## Just for my convenience
        #print(whole1[k][0],"\n")  ## Just for my convenience
        n_neigh_of_o_list.append(whole1[k][0])
        whole1=[]
        
    #print("n_neigh_of_o_list=\t",n_neigh_of_o_list,"\n") #For my convenience
    
    ##### Calculate the reach-distance(p,o); Status- Achieved;
    reach_dist = []
    for i5 in range(len(n_neigh_list)):
        dist21 = distance.euclidean(data[j2],n_neigh_list[i5])
        dist2 = max(n_neigh_of_o_list[i5],dist21)
        reach_dist.append(dist2)
    
    #print("Reach-distance(p,o)=\t",reach_dist,"\n") #For my convenience
    
    ##### Calculate the lrd(Local Reachability Distance) of the datapoint; Status- Achieved;
    sum_lrd = 0.0
    for i5 in reach_dist:
        sum_lrd = sum(reach_dist)
    lrd = k/sum_lrd
    #print(sum_lrd,"\n")
    print("Local Reachability Distance(lrd) of",data[j2],"=","\t",lrd,"\n") 
    
    
    
    ##### Calculate the lrd(Local Reachability Distance) of all k-nearest neighbours of the datapoint; Stataus- Achieved;
    
    tem11 = len(n_neigh_list)
    
    whole_sub = []
    n_neigh_list_sub = []
    
    for j6 in range(0,tem11):
        for i6 in range(0,tem1):                                                      # Changed tem11 to tem1
            dist_sub = distance.euclidean(n_neigh_list[j6],data[i6])                  #Changed n_neigh_list[i6] to data[i6]
            whole_sub.append([dist_sub,n_neigh_list[j6],data[i6]])                    #Changed n_neigh_list[i6] to data[i6]
        whole_sub.sort(key=lambda y: y[0])
        for i7 in range(1,(k+1)):                                                     #Changed k to (k+1):
            n_neigh_list_sub.append(whole_sub[i7][2])
        #print("The",k,"nearest neighbours of",n_neigh_list[j6],"are:\n",n_neigh_list_sub,"\n") #For my convenience
        
        ##### To get the values of the k-distance of 'o'; Status- Achieved; "SUBSTITUTE"
        len_n_neigh_list_sub = len(n_neigh_list_sub)
    
        whole1_sub = []
        n_neigh_of_o_list_sub = []
        for i8 in range(0,len_n_neigh_list_sub):
            for j8 in range(0,tem1):  #Changed tem11 to tem1
                dist1_sub = distance.euclidean(n_neigh_list_sub[i8],data[j8])         #Changed n_neigh_list[j8] to data[j8]
                whole1_sub.append([dist1_sub,n_neigh_list_sub[i8],data[j8]])
            whole1_sub.sort(key=lambda y: y[0])
            n_neigh_of_o_list_sub.append(whole1_sub[k][0])                            
            whole1_sub=[]
        #print("n_neigh_of_o_list=\t",n_neigh_of_o_list_sub,"\n")                     #For my convenience
    
        ##### Calculate the reach-distance(p,o); Status- Achieved;
        reach_dist_sub = []
        for i9 in range(len(n_neigh_list_sub)):
            dist21_sub = distance.euclidean(n_neigh_list[j6],n_neigh_list_sub[i9])
            dist2_sub = max(n_neigh_of_o_list_sub[i9],dist21_sub)
            reach_dist_sub.append(dist2_sub)
        #print("Reach-distance(p,o)=\t",reach_dist_sub,"\n")                          #For my convenience
        
        ##### Calculate the lrd(Local Reachability Distance) of the datapoint; Status- Achieved;
        sum_lrd_sub = 0.0
        for i9 in reach_dist_sub:
            sum_lrd_sub = sum(reach_dist_sub)
        lrd_sub = k/sum_lrd_sub
        #print("Local Reachability Distance(lrd) of",n_neigh_list[j6],"=","\t",lrd_sub,"\n\n") #For my convenience
        
        lrd_of_o_for_LOF.append(lrd_sub) #This stores the lrd of the neighbours of p
        #########################################
        n_neigh_of_o_list_sub = []
        reach_dist_sub = []
        whole_sub=[]
        n_neigh_list_sub=[]
    
    print("\nlrd values of neighbour of",data[j2],"-",lrd_of_o_for_LOF,"\n")
    
    ##### Calculate the LOF value of the datapoint 'p'; Status- Achieved;
    lof_values_sum = 0.0
    for i in lrd_of_o_for_LOF:
        sag2 = i/lrd
        lof_values_sum = lof_values_sum + sag2
    lof_value = lof_values_sum/k
    print("\nLOF score of",data[j2],"=",lof_value,"\n")
    
    ##### Decide whether or not the datapoint 'p' is an outlier or not
    if (lof_value>lof_value_by_me):
        print("\n",data[j2],"is an outlier")
        outlier_list.append(data[j2])
    
    lrd_of_o_for_LOF = []
    n_neigh_of_o_list = []   
    reach_dist = []
    whole=[]
    n_neigh_list=[]

    print("\n*********************************************************************\n")
    ####

print("\n\nOUTLIERS =",outlier_list)
outlier_array = np.array(outlier_list)

##### Plotting for 2D data only
if (int(columns) == 2):
    import matplotlib.pyplot as plt
    plt.figure(figsize=(8,8))
    plt.scatter(data[:,0],data[:,1],c='blue',label='Non-Outliers')
    plt.scatter(outlier_array[:,0],outlier_array[:,1],c='red',s=100,marker='*',label='Outliers')
    plt.plot(label='Outlier')
    plt.xlabel('X-Coordinates')
    plt.ylabel('Y-Coordinates')
    plt.title('Outlier with Non-Outlier visualization')
    plt.legend()
    plt.show()
 
elapsed = time.time() - tim
print("\nTime taken =",elapsed,"sec")
```
